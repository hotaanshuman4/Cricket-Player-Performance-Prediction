```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

br = pd.read_csv('/content/newplayer (1).csv')
br['Opposition'] = br['Opposition'].str.replace('v ', '')

br.columns = [col.replace(" ", "").replace("_", "").strip() for col in br.columns]
br['Player'] = br['Player'].str.strip()
br['Opposition'] = br['Opposition'].str.strip()

demo = br[['Player', 'Opposition', 'BF', 'Overs', 'Runs']].dropna(subset=['BF', 'Overs', 'Runs'])

demo_encoded = pd.get_dummies(demo, columns=['Player', 'Opposition'], prefix=['Player', 'Opposition'])

X = demo_encoded.drop('Runs', axis=1)
Y = demo_encoded['Runs']

train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.20, shuffle=True)

tree = DecisionTreeRegressor()
tree.fit(train_X, train_Y)

print("ðŸ Predicting Runs of a Player Against a Team ðŸ")
player_input = input("Enter player name (e.g., Virat Kohli): ").strip()  # Strip input to match normalized data
opposition_input = input("Enter opposition team (e.g., Pakistan): ").strip()  # Strip input to match normalized data
bf = float(input("Enter balls faced: "))
ov = float(input("Enter overs: "))

player_col = f'Player_{player_input}'
opposition_col = f'Opposition_{opposition_input}'

print(f"\nDebug: Looking for player column: '{player_col}'")
print(f"Debug: Looking for opposition column: '{opposition_col}'")
print(f"Debug: Player column exists: {player_col in X.columns}")
print(f"Debug: Opposition column exists: {opposition_col in X.columns}")

if player_col in X.columns and opposition_col in X.columns:

    input_data = pd.DataFrame(0, index=[0], columns=X.columns)
    input_data.loc[0, player_col] = 1
    input_data.loc[0, opposition_col] = 1
    input_data.loc[0, 'BF'] = bf
    input_data.loc[0, 'Overs'] = ov


    predicted_run = int(tree.predict(input_data)[0])

    print(f"\nâœ… Predicted Runs: {predicted_run}")
    print(f"ðŸ“Œ {player_input} is expected to score {predicted_run} runs against {opposition_input}")
else:
    print("\nâŒ Error: Player or opposition not found in the dataset.")
    print("ðŸ“‹ Available players:", br['Player'].unique().tolist())
    print("ðŸ“‹ Available oppositions:", br['Opposition'].unique().tolist())
```

```python

br = pd.read_csv('/content/newplayer (1).csv')
br['Opposition'] = br['Opposition'].str.replace('v ', '')

br.columns = [col.replace(" ", "").replace("_", "").strip() for col in br.columns]
br['Player'] = br['Player'].str.strip()
br['Opposition'] = br['Opposition'].str.strip()

demo = br[['Player', 'Opposition', 'Overs', 'Runs']].dropna(subset=['Overs', 'Runs'])

demo_encoded = pd.get_dummies(demo, columns=['Player', 'Opposition'], prefix=['Bowler', 'Opposition'])

X = demo_encoded.drop('Runs', axis=1)
Y = demo_encoded['Runs']

train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.20, shuffle=True)

tree = DecisionTreeRegressor()
tree.fit(train_X, train_Y)

# === Prediction for Bowler ===
print("ðŸ Predicting Runs Conceded by a Bowler Against a Team ðŸ")
bowler_input = input("Enter bowler name (e.g., Jasprit Bumrah): ").strip()
opposition_input = input("Enter opposition team (e.g., Pakistan): ").strip()
overs_input = float(input("Enter overs bowled: "))

bowler_col = f'Bowler_{bowler_input}'
opposition_col = f'Opposition_{opposition_input}'

print(f"\nDebug: Looking for bowler column: '{bowler_col}'")
print(f"Debug: Looking for opposition column: '{opposition_col}'")
print(f"Debug: Bowler column exists: {bowler_col in X.columns}")
print(f"Debug: Opposition column exists: {opposition_col in X.columns}")


if bowler_col in X.columns and opposition_col in X.columns:

    input_data = pd.DataFrame(0, index=[0], columns=X.columns)
    input_data.loc[0, bowler_col] = 1
    input_data.loc[0, opposition_col] = 1
    input_data.loc[0, 'Overs'] = overs_input


    predicted_runs_conceded = int(tree.predict(input_data)[0])

    print(f"\nâœ… Predicted Wickets Conceded: {predicted_runs_conceded}")
    print(f"ðŸ“Œ {bowler_input} is expected to concede {predicted_runs_conceded} runs against {opposition_input} in {overs_input} overs")
else:
    print("\nâŒ Error: Bowler or opposition not found in the dataset.")
    print("ðŸ“‹ Available bowlers (players in dataset):", br['Player'].unique().tolist())
    print("ðŸ“‹ Available oppositions:", br['Opposition'].unique().tolist())
```
